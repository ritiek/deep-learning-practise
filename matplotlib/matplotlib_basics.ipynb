```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state to a particular seed for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
# the below multiplies every index in np.random.random(10000) array with (mu+sigma)
x = mu + sigma * np.random.randn(10000) # np.random.randn(10000) means 10000 random numbers in array near to 1

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
# plt.hist(x, no. of bars, normed=1, facecolor=colorLetter, alpha=opacity)

# normed=0 to mark same intensity for all IQ's. see next example

plt.xlabel('Smarts')
plt.ylabel('Probability', fontsize=14)
plt.title('Histogram of IQ', color='red')
# ('myText', fontsize=0, color='black')

plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # = "=100, =15". plt.text(dist. x-axis, dist. y-axis, 'myText')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()

# the only thing i don't get is why wouldn't plt.hist() work correctly if i change x to some custom array
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state to a particular seed for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
# the below multiplies every index in np.random.random(10000) array with (mu+sigma)
x = mu + sigma * np.random.randn(10000) # np.random.randn(10000) means 10000 random numbers in array near to 1

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
# plt.hist(x, no. of bars, normed=1, facecolor=colorLetter, alpha=opacity)

# normed=0 to mark same intensity for all IQ's. see next example

plt.xlabel('Smarts')
plt.ylabel('Probability', fontsize=14)
plt.title('Histogram of IQ', color='red')
# ('myText', fontsize=0, color='black')

plt.text(60, .025, r'$\mu=100,\ \sigma=15$', fontsize=11) # = "=100, =15". plt.text(dist. x-axis, dist. y-axis, 'myText')
plt.axis([40, 160, 0, 0.035])
plt.grid(True)
plt.annotate('Average IQ', color='magenta', xy=(100, 0.028), xytext=(120, 0.032),
            arrowprops=dict(facecolor='orange', shrink=0.1),
            )
plt.show()

# the only thing i don't get is why wouldn't plt.hist() work correctly if i change x to some custom array
```
